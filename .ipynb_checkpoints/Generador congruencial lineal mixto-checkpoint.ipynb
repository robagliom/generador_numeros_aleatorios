```python
#Generar los valores U1 a U20 para identificar un comportamiento significativo (Z0=7, m=16, a=5, c=3)
Z0 = 7
m = 16
a = 5
c = 3
```

```python
def generador_congruencial(zi_menos1, m, a, c):
    zi = (a*zi_menos1+c) % m
    return zi
```

```python
zi_menos1 = Z0
for i in range(1,21):
    Zi = generador_congruencial(zi_menos1,m,a,c)
    zi_menos1 = Zi
    Ui = Zi/m
    print('U{}: '.format(i),Ui)
```

```python
def divisores(numero):
    divisor = 0
    divisores = []

    if numero % 2 == 0:
        iterar = numero / 2
    else:
        iterar = (numero - 1) / 2

    for i in range(1, int(iterar) + 1):
        if numero % i == 0:
            aux = numero / i
            if aux != divisor:
                divisor = aux
            divisores.append(int(divisor))
    divisores.append(1)
    return(divisores)

def interseccion_listas(a, b):
    lista_final = []
    for i in a:
        if (i not in lista_final) and (i in b):
            lista_final.append(i)
    return lista_final
```

```python
#Vamos a verificar si se cumple el
#teorema: dado un GCL mixto se puede saber si es de período completo si se cumple lo siguiente
#1- El único entero que divide a m y c en forma simultanea (primos entre si) es 1.
#2- Si q es un número primo que divide a m, entonces q divide a a-1
#3- Si 4 divide a m, entonces 4 divide a a -1

#1-
print("# Punto 1")
divisores_m = divisores(m)
divisores_c = divisores(c)
print("Divisores de m = {}:".format(m),divisores_m)
print("Divisores de c = {}:".format(c),divisores_c)
divisores_comunes = interseccion_listas(divisores_m, divisores_c)
print("Divisores en común: ",divisores_comunes)
if divisores_comunes == [1]:
    print("Se cumple que el único que divide a m y c en forma simultanea (primos entre si) es 1.\n")
    
#2-
print("#Punto 2")
print("Hipóstesis: q primo (divisible por q y por 1). q divide a m.")
print("Quiero probar: q divide a a-1.")
print("Tomamos q = 1 y q = 2, únicos número primo que divide a m")
q = 1
print("Para q = 1: m/q:",m%q,' --> q divide a m')
print("Luego, (a-1)/q:", (a-1)%q,' por lo tanto q divide a a-1')
q = 2
print("Para q = 2: m/q:",m%q,' --> q divide a m')
print("Luego, (a-1)/q:", (a-1)%q,' por lo tanto q divide a a-1\n')

#3-
print("#Punto 3")
print("Hipóstesis: 4 divide a m.")
print("Quiero probar: 4 divide a a -1.")
if 4 in divisores_m:
    print("4 divide a m")
    print("Ahora veamos si 4 divide a a-1:", (a-1)%4, ' --> 4 divide a a-1\n')
    
#Conclusión
print("Por lo tanto, se cumple el teorema")
```
